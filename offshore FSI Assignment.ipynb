```python
from scipy.special import j1,y1
import numpy as np
import matplotlib.pyplot as plt
import math

def af(ka):
    return 1/(np.sqrt(j1(ka)**2+y1(ka)**2))

L_D = [1,2,4,8,16]
x=[]
y=[]

for i in range(len(L_D)):
    ka=2*np.pi*pow(L_D[i],-1)
    #print(ka)
    CM= (4*af(ka))/(np.pi*ka**2)
    x.append(CM)
    y.append(af(ka))

#print(x)
#print(y)
plt.plot(x, y,label=L_D)
plt.legend(title='$L/D$')
plt.xlabel('Cm')
plt.ylabel('A_function',rotation=0)
plt.title('Cm vs af(ka)')
plt.show()
print(x)
```
